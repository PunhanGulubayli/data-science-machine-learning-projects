```python
final_df = final_df.merge(recency, on = 'CustomerID', how = 'inner').copy()
```

```python
final_df = final_df[final_df['monetary'] > 0].copy()
```

```python
def categorizeFreq(col):
    if col < 10: 
        return "freq = Worst"
    if col < 30:
        return "freq = Bad"
    if col < 50 :
        return "freq = Good"
    return "freq = Best"


def categorizeMon(col):
    if col < 100: 
        return "mon = Worst"
    if col < 300:
        return "mon = Bad"
    if col < 500 :
        return "mon = Good"
    return "mon = Best"


def categorizeRec(col):
    if col < 30: 
        return "rec = Best"
    if col < 60:
        return "rec = Good"
    if col < 90 :
        return "rec = Bad"
    return "rec = Worst"
```

```python
final_df['fCat'] = final_df['freq'].apply(categorizeFreq)
final_df['rCat'] = final_df['recency'].apply(categorizeRec)
final_df['mCat'] = final_df['monetary'].apply(categorizeMon)
```

```python
final_df['cluster'] = final_df['fCat'] + ' ' + final_df['rCat'] + ' ' + final_df['mCat']
```

```python
import plotly.express as px
fig = px.scatter_3d(final_df, x='recency', y='freq', z='monetary',
              color='cluster')
fig.write_html('result.html')
```
