```python
dataset = df[['CustomerID', 'InvoiceDate', 'UnitPrice', 'Quantity']].copy()
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace = True)
```

```python
dataset
```

```python
dataset['CustomerID'] = dataset['CustomerID'].apply(lambda x : int(x))
```

```python
dataset.dtypes
```

```python
dataset['Amount'] = dataset['UnitPrice'] * dataset['Quantity']
```

```python
dataset
```

```python
dataset.drop(['UnitPrice', 'Quantity'], axis = 1, inplace = True)
```

```python
dataset
```

```python
dataset['InvoiceDate'] = pd.to_datetime(dataset['InvoiceDate']).dt.date
```

```python
dataset
```

```python
dataset.groupby('CustomerID').nunique()[['InvoiceDate']]
```

```python
final_df = dataset.groupby('CustomerID').agg({
    "InvoiceDate" : "nunique",
    "Amount" : "mean"})
```

```python
final_df
```

```python
final_df.rename({"InvoiceDate" : "Frequency", "Amount" : "Monetary"}, axis = 1, inplace = True)
```

```python
final_df
```

```python
final_df.reset_index(inplace = True)
```

```python
recency = dataset.InvoiceDate.max() - dataset.groupby('CustomerID').max()[['InvoiceDate']]
recency['InvoiceDate'] = recency['InvoiceDate'].apply(lambda x : x.days)
```

```python
recency.rename({"InvoiceDate" : "Recency"}, axis =1, inplace = True)
```

```python
recency.reset_index(inplace = True)
```

```python
recency
```

```python
final_df = final_df.merge(recency, on = 'CustomerID', how = 'inner').copy()
```

```python
final_df
```

```python
final_df = final_df[final_df['Monetary'] > 0].copy()
```

```python
def categorizeFreq(col):
    if col < 10: 
        return "freq = Worst"
    if col < 30:
        return "freq = Bad"
    if col < 50 :
        return "freq = Good"
    return "freq = Best"


def categorizeMon(mon):
    if mon < 100: 
        return "mon = Worst"
    if mon < 300:
        return "mon = Bad"
    if mon < 500 :
        return "mon = Good"
    return "mon = Best"


def categorizeRec(rec):
    if rec < 30: 
        return "rec = Best"
    if rec < 60:
        return "rec = Good"
    if rec < 90 :
        return "rec = Bad"
    return "rec = Worst"
```

```python
final_df['Frequency Category'] = final_df['Frequency'].apply(categorizeFreq)
final_df['Recency Category'] = final_df['Recency'].apply(categorizeRec)
final_df['Monetary Category'] = final_df['Monetary'].apply(categorizeMon)
```

```python
final_df['Cluster'] = final_df['Frequency Category'] + " " +final_df['Recency Category'] + " " + final_df['Monetary Category']
```

```python
final_df
```

```python
import plotly.express as px
fig = px.scatter_3d(final_df, x='Recency', y='Frequency', z='Monetary',
              color='Cluster')
fig.write_html('result.html')
```
